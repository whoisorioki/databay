```python
previously_ranked = df[df["previous_rank"].notnull()]
rank_change = previously_ranked["previous_rank"] - previously_ranked["rank"]
df["rank_change"] = rank_change
df.head()
```

```python
big_rev_neg_profit = df[(df["revenues"] > 100000) & (df["profits"] < 0)]
```

```python
brazil_venezuela = df[(df['country'] == 'Brazil') | (df['country'] == 'Venezuela')]
tech_outside_usa = df[(df['country'] != 'USA') & (df['sector'] == 'Technology')]
```

```python
selected_rows = df[df["country"] == "Japan"]
sorted_rows = selected_rows.sort_values("employees", ascending=False)
top_japanese_employer = sorted_rows.iloc[0]["company"]
top_japanese_employer
```

```python
avg_rev_by_country = {}
countries = df["country"].unique()

for c in countries:
    selected_rows = df[df["country"] == c]
    mean = selected_rows["revenues"].mean()
    avg_rev_by_country[c] = mean

avg_rev_by_country
```

```python
top_employer_by_country = {}

for country in df["country"].unique():
    selected_rows = df[df["country"] == country]
    sorted_rows = selected_rows.sort_values("employees", ascending=False)
    top_employer = sorted_rows.iloc[0]
    employer_name = top_employer["company"]
    top_employer_by_country[country] = employer_name

top_employer_by_country
```

```python
df["roa"] = df["profits"] / df["assets"]

top_roa_by_sector = {}
for sector in df["sector"].unique():
    is_sector = df["sector"] == sector
    sector_companies = df.loc[is_sector]
    top_company = sector_companies.sort_values("roa",ascending=False).iloc[0]
    company_name = top_company["company"]
    top_roa_by_sector[sector] = company_name

top_roa_by_sector
```
