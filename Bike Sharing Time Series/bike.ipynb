```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load data
df = pd.read_csv('day.csv')
df.tail()
```

```python
# Inspecting the dataframe
df.info()
```

```python
df.duplicated().sum()
```

```python
df['dteday'] = pd.to_datetime(df['dteday'])
```

```python
# Random plot
plt.plot(df['dteday'], df['cnt'])
plt.xticks(rotation=75)
plt.ylabel('Bikes Rented')
plt.xlabel('Date')
plt.show()
```

```python
plt.plot(df['dteday'], df['casual'], label='Casual')
plt.plot(df['dteday'], df['registered'], label='Registered')
plt.xticks(rotation=50)
plt.ylabel('Bikes Rented')
plt.xlabel('Date')
plt.title('Bikes Rented: Casual vs. Registered')
plt.legend()
plt.show()
```
